```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/foodhub_order.csv') # Use pd.read_csv to read the file

# Print the column names to verify if 'food_preperation_time' exists and check for typos.
print(data.columns)

plt.figure(figsize=(10,6))

# If the column name is misspelled, correct it here. For example, if the correct name is 'food_preparation_time':
sns.boxplot(data=data,x='cuisine_type',y='food_preparation_time', showfliers=False) # Use showfliers to hide outlier points

plt.xticks(rotation=90)
plt.show()
```

```python
sns.histplot(data=data,x='delivery_time',y='rating')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=data,x='day_of_the_week',y='order_id',hue='cuisine_type')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(data.corr(),annot=True,cmap='spectrol',Vmin=-1,Vmax=1)

plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

# Select only numerical columns for correlation calculation
numerical_data = data.select_dtypes(include=np.number)

sns.heatmap(numerical_data.corr(),annot=True,cmap='Spectral',vmin=-1,vmax=1)

plt.xticks(rotation=90)
plt.show()
```
